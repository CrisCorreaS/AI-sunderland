```python
from collections import deque


def bfs(graph, start, goal):
    visited = set()
    queue = deque([(start, [])])

    while queue:
        node, path = queue.popleft()
        print(f"Current node: {node}, Path: {path + [node]}")

        if node == goal:
            return path + [node]

        if node not in visited:
            visited.add(node)

            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append((neighbor, path + [node]))

    print("Goal not found")
    return None
```

```python
# From Longbenton (60) to Gateshead (16)

start_node1 = "60"
goal_node1 = "16"

result1 = bfs(graph, start_node1, goal_node1)
print()

if result1:
    print("-> Path found:", result1)
else:
    print("-> No path found.")
```
